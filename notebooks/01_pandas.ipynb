# ETL con Pandas - Market Data

## Importamos las librerías necesarias

```python
import pandas as pd
import numpy as np
import re

print("✅ Librerías importadas")
```

## EXTRACT: Cargar los datos

```python
df_raw = pd.read_csv("../data/market_data.csv")

print(f"✅ Dataset cargado")
print(f"Filas: {df_raw.shape[0]}")
print(f"Columnas: {df_raw.shape[1]}")
print(f"\nColumnas: {list(df_raw.columns)}")
df_raw.head(10)
```

## EXPLORACIÓN: Análisis del dataset

```python
print("=== INFORMACIÓN GENERAL ===\n")
print(f"Dimensiones: {df_raw.shape[0]} filas × {df_raw.shape[1]} columnas")
print(f"\nTipos de datos:")
print(df_raw.dtypes)
```

```python
print("=== VALORES NULOS ===\n")
print(df_raw.isnull().sum())
print(f"\nTotal nulos: {df_raw.isnull().sum().sum()}")
```

```python
print("=== DUPLICADOS ===\n")
print(f"Filas duplicadas: {df_raw.duplicated().sum()}")
```

```python
print("=== VALORES ÚNICOS POR COLUMNA ===\n")
for col in df_raw.columns:
    print(f"{col}: {df_raw[col].nunique()}")
```

```python
print("=== VALORES ÚNICOS EN COLUMNAS CATEGÓRICAS ===\n")

print("CURRENCY:")
print(df_raw['currency'].unique())

print("\nSECTOR:")
print(df_raw['sector'].unique())

print("\nMARKET_TYPE:")
print(df_raw['market_type'].unique())

print("\nEXCHANGE:")
print(df_raw['exchange'].unique())

print("\nCOUNTRY:")
print(df_raw['country'].unique())

print("\nNAME:")
print(df_raw['name'].unique())

print("\nSYMBOL:")
print(df_raw['symbol'].unique())
```

```python
print("=== FORMATOS DE FECHA ===\n")
print("Ejemplos de 'date':")
print(df_raw['date'].head(20).values)
```

```python
print("=== VALORES PROBLEMÁTICOS ===\n")

valores_problematicos = ['?', 'N/A', 'n/a', 'NULL', 'null', 'Unknown', 'unknown', '', ' ', '-']

for col in df_raw.columns:
    count = df_raw[col].isin(valores_problematicos).sum()
    if count > 0:
        print(f"{col}: {count} valores problemáticos")
```

```python
print("=== COLUMNAS CON TEXTO DONDE DEBERÍA HABER NÚMEROS ===\n")

cols_numericas = ['open', 'close', 'high', 'low', 'volume', 'market_cap']

for col in cols_numericas:
    if df_raw[col].dtype == 'object':
        con_texto = df_raw[col].astype(str).str.contains('[a-zA-Z]', na=False).sum()
        print(f"{col}: {con_texto} valores con texto")
        if con_texto > 0:
            ejemplos = df_raw[df_raw[col].astype(str).str.contains('[a-zA-Z]', na=False)][col].head(3).values
            print(f"   Ejemplos: {ejemplos}")
```
