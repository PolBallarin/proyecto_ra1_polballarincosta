## TRANSFORM: Limpieza de datos

### Paso 1: Crear copia de trabajo

```python
df_clean = df_raw.copy()
print(f"✅ Copia creada: {len(df_clean)} filas")
```

### Paso 2: Limpiar columna 'open' (extraer números)

```python
def extract_number(value):
    if pd.isna(value):
        return pd.NA
    value_str = str(value)
    numbers = re.findall(r'[\d.]+', value_str)
    if numbers:
        return float(numbers[0])
    return pd.NA

df_clean['open'] = df_clean['open'].apply(extract_number)

print(f"✅ 'open' limpiado")
print(f"Tipo: {df_clean['open'].dtype}")
print(f"Nulos: {df_clean['open'].isnull().sum()}")
```

### Paso 3: Normalizar columna 'currency'

```python
print("Antes:", df_clean['currency'].unique())

df_clean['currency'] = df_clean['currency'].str.upper().replace({
    '$': 'USD',
    'USDT': 'USD'
})

print("Después:", df_clean['currency'].unique())
print("✅ 'currency' normalizado")
```

### Paso 4: Normalizar columna 'sector'

```python
print("Antes:", df_clean['sector'].unique())

sector_map = {
    'finance': 'Finance',
    'Finance': 'Finance',
    'tech': 'Technology',
    'Tech': 'Technology',
    'TECHNOLOGY': 'Technology',
    'Technology': 'Technology',
    'retail': 'Retail',
    'RETAIL': 'Retail',
    'Retail': 'Retail',
    'crypto': 'Crypto',
    'Crypto': 'Crypto',
    'blockchain': 'Crypto',
    'Blockchain': 'Crypto',
    'ai': 'AI',
    'AI': 'AI',
    'Artificial Intelligence': 'AI',
    'automotive': 'Automotive',
    'Automotive': 'Automotive'
}

df_clean['sector'] = df_clean['sector'].replace(sector_map)

print("Después:", df_clean['sector'].unique())
print("✅ 'sector' normalizado")
```

### Paso 5: Limpiar columna 'exchange'

```python
print("Antes:", df_clean['exchange'].unique())
print(f"Nulos: {df_clean['exchange'].isnull().sum()}")
print(f"'Unknown': {(df_clean['exchange'] == 'Unknown').sum()}")

# Reemplazar 'Unknown' por NaN, luego rellenar todos los NaN
df_clean['exchange'] = df_clean['exchange'].replace('Unknown', pd.NA)
df_clean['exchange'] = df_clean['exchange'].fillna('exchange_empty')

print("\nDespués:", df_clean['exchange'].unique())
print("✅ 'exchange' limpiado")
```

### Paso 6: Limpiar columna 'symbol'

```python
print(f"Nulos en 'symbol': {df_clean['symbol'].isnull().sum()}")

# Crear mapeo name → symbol usando filas que sí tienen symbol
name_to_symbol = df_clean[df_clean['symbol'].notna()].drop_duplicates('name').set_index('name')['symbol'].to_dict()

# Rellenar nulos usando el mapeo
df_clean['symbol'] = df_clean.apply(
    lambda row: name_to_symbol.get(row['name'], 'symbol_empty') if pd.isna(row['symbol']) else row['symbol'],
    axis=1
)

print(f"Nulos después: {df_clean['symbol'].isnull().sum()}")
print(f"'symbol_empty' restantes: {(df_clean['symbol'] == 'symbol_empty').sum()}")
print("✅ 'symbol' limpiado")
```

### Paso 7: Normalizar columna 'date' (3 formatos distintos)

```python
print("Ejemplos antes:")
print(df_clean['date'].head(10).values)

def parse_date(value):
    if pd.isna(value):
        return pd.NA
    
    value_str = str(value).strip()
    
    # Formato YYYY-MM-DD (2023-06-13)
    if re.match(r'^\d{4}-\d{2}-\d{2}$', value_str):
        return pd.to_datetime(value_str, format='%Y-%m-%d').date()
    
    # Formato DD/MM/YYYY (13/08/2024)
    if re.match(r'^\d{2}/\d{2}/\d{4}$', value_str):
        return pd.to_datetime(value_str, format='%d/%m/%Y').date()
    
    # Formato MM-DD-YYYY (08-03-2024)
    if re.match(r'^\d{2}-\d{2}-\d{4}$', value_str):
        return pd.to_datetime(value_str, format='%m-%d-%Y').date()
    
    return pd.NA

df_clean['date'] = df_clean['date'].apply(parse_date)

print("\nEjemplos después:")
print(df_clean['date'].head(10).values)
print(f"\nTipo: {df_clean['date'].dtype}")
print(f"Nulos: {df_clean['date'].isnull().sum()}")
print("✅ 'date' normalizado")
```

### Paso 8: Convertir 'last_updated' a datetime

```python
print(f"Tipo antes: {df_clean['last_updated'].dtype}")

df_clean['last_updated'] = pd.to_datetime(df_clean['last_updated'])

print(f"Tipo después: {df_clean['last_updated'].dtype}")
print("✅ 'last_updated' convertido")
```

### Paso 9: Verificación final

```python
print("=== RESUMEN DE LA TRANSFORMACIÓN ===\n")

print(f"Filas: {len(df_clean)}")
print(f"Columnas: {len(df_clean.columns)}")

print(f"\n--- Tipos de datos ---")
print(df_clean.dtypes)

print(f"\n--- Valores nulos ---")
print(df_clean.isnull().sum())

print(f"\n--- Valores únicos en columnas categóricas ---")
for col in ['currency', 'sector', 'exchange', 'symbol', 'market_type', 'country']:
    print(f"{col}: {df_clean[col].unique()}")
```

```python
print("=== COMPARACIÓN ANTES / DESPUÉS ===\n")

print("ANTES (df_raw):")
print(df_raw[['open', 'currency', 'sector', 'exchange', 'date']].head())

print("\nDESPUÉS (df_clean):")
print(df_clean[['open', 'currency', 'sector', 'exchange', 'date']].head())
```

### Paso 10: Crear columnas derivadas

```python
# Cambio diario (absoluto)
df_clean['daily_change'] = df_clean['close'] - df_clean['open']

# Cambio diario (porcentaje)
df_clean['daily_change_pct'] = ((df_clean['close'] - df_clean['open']) / df_clean['open']) * 100

print("✅ Columnas derivadas creadas")
print(df_clean[['open', 'close', 'daily_change', 'daily_change_pct']].head(10))
```

### Paso 11: Reducir decimales

```python
cols_decimales = ['open', 'close', 'high', 'low', 'volume', 'market_cap', 'daily_change', 'daily_change_pct']

for col in cols_decimales:
    df_clean[col] = df_clean[col].round(2)

print("✅ Decimales reducidos a 2 dígitos")
print(df_clean[['open', 'close', 'daily_change', 'daily_change_pct']].head())
```

## LOAD: Guardar datos limpios

### Paso 12: Guardar CSV limpio

```python
df_clean.to_csv("../data/market_data_clean.csv", index=False)

print("✅ CSV limpio guardado en: ../data/market_data_clean.csv")
print(f"Filas: {len(df_clean)}")
print(f"Columnas: {len(df_clean.columns)}")
```

### Paso 13: Crear tablas dimensionales

```python
# dim_asset (symbol, name, market_type)
df_dim_asset = df_clean[['symbol', 'name', 'market_type']].drop_duplicates().reset_index(drop=True)
df_dim_asset['asset_id'] = df_dim_asset.index + 1
df_dim_asset = df_dim_asset[['asset_id', 'symbol', 'name', 'market_type']]
print(f"✅ dim_asset: {len(df_dim_asset)} registros")
print(df_dim_asset)
```

```python
# dim_sector
df_dim_sector = pd.DataFrame({'sector_name': df_clean['sector'].unique()})
df_dim_sector['sector_id'] = df_dim_sector.index + 1
df_dim_sector = df_dim_sector[['sector_id', 'sector_name']]
print(f"✅ dim_sector: {len(df_dim_sector)} registros")
print(df_dim_sector)
```

```python
# dim_exchange
df_dim_exchange = pd.DataFrame({'exchange_name': df_clean['exchange'].unique()})
df_dim_exchange['exchange_id'] = df_dim_exchange.index + 1
df_dim_exchange = df_dim_exchange[['exchange_id', 'exchange_name']]
print(f"✅ dim_exchange: {len(df_dim_exchange)} registros")
print(df_dim_exchange)
```

```python
# dim_currency
df_dim_currency = pd.DataFrame({'currency_name': df_clean['currency'].unique()})
df_dim_currency['currency_id'] = df_dim_currency.index + 1
df_dim_currency = df_dim_currency[['currency_id', 'currency_name']]
print(f"✅ dim_currency: {len(df_dim_currency)} registros")
print(df_dim_currency)
```

```python
# dim_country
df_dim_country = pd.DataFrame({'country_name': df_clean['country'].unique()})
df_dim_country['country_id'] = df_dim_country.index + 1
df_dim_country = df_dim_country[['country_id', 'country_name']]
print(f"✅ dim_country: {len(df_dim_country)} registros")
print(df_dim_country)
```

```python
# dim_date
df_dim_date = pd.DataFrame({'date': df_clean['date'].unique()})
df_dim_date['date_dt'] = pd.to_datetime(df_dim_date['date'])
df_dim_date['year'] = df_dim_date['date_dt'].dt.year
df_dim_date['month'] = df_dim_date['date_dt'].dt.month
df_dim_date['day'] = df_dim_date['date_dt'].dt.day
df_dim_date['weekday'] = df_dim_date['date_dt'].dt.day_name()
df_dim_date['date_id'] = df_dim_date.index + 1
df_dim_date = df_dim_date[['date_id', 'date', 'year', 'month', 'day', 'weekday']]
print(f"✅ dim_date: {len(df_dim_date)} registros")
print(df_dim_date.head(10))
```

### Paso 14: Crear tabla de hechos (fact_market_data)

```python
# Crear copia para la tabla de hechos
df_fact = df_clean.copy()

# Añadir FKs desde las dimensiones
df_fact = df_fact.merge(df_dim_asset[['asset_id', 'symbol']], on='symbol', how='left')
df_fact = df_fact.merge(df_dim_sector[['sector_id', 'sector_name']], left_on='sector', right_on='sector_name', how='left')
df_fact = df_fact.merge(df_dim_exchange[['exchange_id', 'exchange_name']], left_on='exchange', right_on='exchange_name', how='left')
df_fact = df_fact.merge(df_dim_currency[['currency_id', 'currency_name']], left_on='currency', right_on='currency_name', how='left')
df_fact = df_fact.merge(df_dim_country[['country_id', 'country_name']], left_on='country', right_on='country_name', how='left')
df_fact = df_fact.merge(df_dim_date[['date_id', 'date']], on='date', how='left')

# Seleccionar solo columnas necesarias
df_fact_market_data = df_fact[[
    'asset_id', 'date_id', 'sector_id', 'exchange_id', 'currency_id', 'country_id',
    'open', 'close', 'high', 'low', 'volume', 'market_cap',
    'daily_change', 'daily_change_pct', 'last_updated'
]].copy()

# Añadir ID de la tabla de hechos
df_fact_market_data.insert(0, 'fact_id', range(1, len(df_fact_market_data) + 1))

print(f"✅ fact_market_data: {len(df_fact_market_data)} registros")
print(f"Columnas: {list(df_fact_market_data.columns)}")
print(df_fact_market_data.head())
```

### Paso 15: Generar DDL y cargar en SQLite

```python
import sqlite3
from sqlalchemy import create_engine

print("=== CARGA EN SQLITE: CREANDO DATAWAREHOUSE ===\n")

# Ruta de la base de datos SQLite
db_path = "../warehouse/warehouse_pandas.db"

# Crear conexión
engine = create_engine(f'sqlite:///{db_path}', echo=False)

# Guardar tablas dimensionales
print("💾 Guardando tablas dimensionales...")
df_dim_asset.to_sql('dim_asset', engine, if_exists='replace', index=False)
df_dim_sector.to_sql('dim_sector', engine, if_exists='replace', index=False)
df_dim_exchange.to_sql('dim_exchange', engine, if_exists='replace', index=False)
df_dim_currency.to_sql('dim_currency', engine, if_exists='replace', index=False)
df_dim_country.to_sql('dim_country', engine, if_exists='replace', index=False)
df_dim_date.to_sql('dim_date', engine, if_exists='replace', index=False)
print("✅ Tablas dimensionales guardadas")

# Guardar tabla de hechos
print("\n💾 Guardando tabla de hechos...")
df_fact_market_data.to_sql('fact_market_data', engine, if_exists='replace', index=False)
print("✅ Tabla de hechos guardada")

print(f"\n✅ Datawarehouse creado en: {db_path}")
print(f"   • 1 tabla de hechos: fact_market_data ({len(df_fact_market_data)} registros)")
print(f"   • 6 tablas dimensionales: dim_asset, dim_sector, dim_exchange, dim_currency, dim_country, dim_date")
```

```python
# Generar DDL desde los DataFrames
def generate_ddl(df, table_name, primary_key=None, foreign_keys=None):
    """Genera DDL basado en un DataFrame"""
    type_map = {
        'int64': 'INTEGER',
        'Int64': 'INTEGER',
        'float64': 'REAL',
        'object': 'TEXT',
        'datetime64[ns]': 'TEXT',
        'bool': 'INTEGER'
    }
    
    columns = []
    for col in df.columns:
        dtype = str(df[col].dtype)
        sql_type = type_map.get(dtype, 'TEXT')
        
        if col == primary_key:
            columns.append(f"    {col} {sql_type} PRIMARY KEY")
        else:
            columns.append(f"    {col} {sql_type}")
    
    ddl = f"CREATE TABLE IF NOT EXISTS {table_name} (\n"
    ddl += ",\n".join(columns)
    
    if foreign_keys:
        for fk_col, ref_table, ref_col in foreign_keys:
            ddl += f",\n    FOREIGN KEY ({fk_col}) REFERENCES {ref_table}({ref_col})"
    
    ddl += "\n);\n"
    return ddl

# Generar DDL completo
ddl_content = """-- ============================================
-- DDL para Datawarehouse - Market Data
-- Generado desde ETL con Pandas
-- ============================================

-- TABLAS DIMENSIONALES

"""

ddl_content += generate_ddl(df_dim_asset, 'dim_asset', 'asset_id')
ddl_content += "\n" + generate_ddl(df_dim_sector, 'dim_sector', 'sector_id')
ddl_content += "\n" + generate_ddl(df_dim_exchange, 'dim_exchange', 'exchange_id')
ddl_content += "\n" + generate_ddl(df_dim_currency, 'dim_currency', 'currency_id')
ddl_content += "\n" + generate_ddl(df_dim_country, 'dim_country', 'country_id')
ddl_content += "\n" + generate_ddl(df_dim_date, 'dim_date', 'date_id')

ddl_content += "\n-- TABLA DE HECHOS\n\n"
ddl_content += generate_ddl(df_fact_market_data, 'fact_market_data', 'fact_id', [
    ('asset_id', 'dim_asset', 'asset_id'),
    ('date_id', 'dim_date', 'date_id'),
    ('sector_id', 'dim_sector', 'sector_id'),
    ('exchange_id', 'dim_exchange', 'exchange_id'),
    ('currency_id', 'dim_currency', 'currency_id'),
    ('country_id', 'dim_country', 'country_id')
])

# Guardar DDL
with open('../warehouse/modelo_datawarehouse_pandas.sql', 'w') as f:
    f.write(ddl_content)

print("✅ DDL guardado en: warehouse/modelo_datawarehouse_pandas.sql")
print("\n=== CONTENIDO DDL ===\n")
print(ddl_content)
```
